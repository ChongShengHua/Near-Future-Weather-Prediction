# Prepare weather type
- make them into 5 type : sunny, cloudy, snowy, foggy, other, rainy

```
df.Icon.unique()
```

```
df = df[df['Icon'] == df['Icon']]
df.Icon.unique()
```

```
df['type'] = ''

for i in range(len(df)):
    if df['Icon'].iloc[i] == 'clear':
        df['type'].iloc[i] = 'sunny'

    elif 'cloudy' in df['Icon'].iloc[i]:
        df['type'].iloc[i] = 'cloudy'
        
    elif 'snow' in df['Icon'].iloc[i]:
        df['type'].iloc[i] = 'snowy'
        
    elif ('fog' in df['Icon'].iloc[i]) or ('sleet' in df['Icon'].iloc[i]) or ('hazy' in df['Icon'].iloc[i]):
        df['type'].iloc[i] = 'foggy'
        
    elif ('rain' in df['Icon'].iloc[i]) or ('tstorms' in df['Icon'].iloc[i]):
        df['type'].iloc[i] = 'rainy'
        
    else:
        df['type'].iloc[i] = 'other'
            
df 
```

# 3.0 Prepare Dataset

```
def cont_img(df, size):
    if size == 3:
        new_df = pd.DataFrame(columns=['CamId', 'img1','img2','img3', 'temp1', 'temp2', 'temp3', 'weather1', 'weather2', 'weather3'])
        date = pd.DataFrame({'Count':df.groupby(['Year','Month','Day']).size()}).reset_index()

        for i in range(len(date)):
            temp = df[((df['Year']==date.Year.iloc[i]) & (df['Month']==date.Month.iloc[i]) &  (df['Day']==date.Day.iloc[i]))].sort_values('Hour')
            for j in range(len(temp)):
                if j+2 < len(temp):
                    new_df.loc[len(new_df)] = [temp.CamId.iloc[j], temp.Filename.iloc[j], temp.Filename.iloc[j+1], temp.Filename.iloc[j+2], temp.TempM.iloc[j], temp.TempM.iloc[j+1], temp.TempM.iloc[j+2], temp['type'].iloc[j], temp['type'].iloc[j+1], temp['type'].iloc[j+2]]            
    
    if size == 4:
        new_df = pd.DataFrame(columns=['CamId', 'img1','img2','img3','img4', 'temp1', 'temp2', 'temp3', 'temp4'])
        date = pd.DataFrame({'Count':df.groupby(['Year','Month','Day']).size()}).reset_index()

        for i in range(len(date)):
            temp = df[((df['Year']==date.Year.iloc[i]) & (df['Month']==date.Month.iloc[i]) &  (df['Day']==date.Day.iloc[i]))].sort_values('Hour')
            for j in range(len(temp)):
                if j+3 < len(temp):
                    new_df.loc[len(new_df)] = [temp.CamId.iloc[j], temp.Filename.iloc[j], temp.Filename.iloc[j+1], temp.Filename.iloc[j+2], temp.Filename.iloc[j+3], temp.TempM.iloc[j], temp.TempM.iloc[j+1], temp.TempM.iloc[j+2], temp.TempM.iloc[j+3], temp['type'].iloc[j], temp['type'].iloc[j+1], temp['type'].iloc[j+2], temp['type'].iloc[j+3]]            
    
    return new_df
```
